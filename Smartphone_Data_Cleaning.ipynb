```python
#importing required libraries:

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading the csv:

df = pd.read_csv("Smartphone.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
# looking at the dataframes summary:
df.info()
```

```python
# removing the string components like comma ',' and texts:
 
df["MRP"] = df["MRP"].str.replace("","")
df["MRP"] = df["MRP"].str.replace(",","")
df["Discount_Price"] = df["Discount_Price"].str.replace("","")
df["Discount_Price"] = df["Discount_Price"].str.replace(",","")
```

```python
# spliting the Reviews data into 2 new columns i.e.total_Rating and total_Reviews:

df[['Total_Rating','Total_Reviews']] = df["Reviews"].str.split('&',expand = True)
```

```python
df.head()
```

```python
# removing Reviews data from the dataframe:

df = df.drop('Reviews',axis = 1)
```

```python
df.head()
```

```python
# Removing string components from Total_Rating and Total_Reviews:

df["Total_Rating"] = df["Total_Rating"].str.replace(" Ratings","")
df["Total_Reviews"] = df["Total_Reviews"].str.replace(" Reviews","")
df["Total_Rating"] = df["Total_Rating"].str.replace(",","")
df["Total_Reviews"] = df["Total_Reviews"].str.replace(",","")
```

```python
df.head()
```

```python
# removing the empty spaces from the data:

df['Total_Rating'] = df['Total_Rating'].str.strip()
df['Total_Reviews'] = df["Total_Reviews"].str.strip()
```

```python
# transforming object datatype into numeric data:

df["MRP"]=pd.to_numeric(df["MRP"])
df["Discount_Price"]=pd.to_numeric(df["Discount_Price"])
df['Total_Rating'] = pd.to_numeric(df["Total_Rating"])
df['Total_Reviews'] = pd.to_numeric(df['Total_Reviews'])
```

```python
# checking the summary again.
df.info()
```

```python
df.to_csv("Smartphone_Cleaned_Data.csv", sep = ',', index = False,encoding = 'UTF-8')
pd.read_csv("Smartphone_Cleaned_Data.csv")
```
