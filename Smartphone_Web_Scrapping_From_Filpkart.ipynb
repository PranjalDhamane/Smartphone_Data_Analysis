### importing libraries

```python
from bs4 import BeautifulSoup
import csv
import requests
import pandas as pd
```

```python
# geting the required urls
url = 'https://www.flipkart.com/search?sid=tyy%2C4io&otracker=CLP_Filters&p%5B%5D=facets.brand%255B%255D%3DAPPLE'
apple_page = requests.get(url)
```

```python
soup = BeautifulSoup(apple_page.content,'html.parser')
```

```python
# extracting apple's data from flipkart

apple_name = soup.find_all('div',attrs='_4rR01T')
apple_mrp = soup.find_all('div',attrs='_3I9_wc _27UcVY')
apple_discount_price = soup.find_all('div',attrs='_30jeq3 _1_WHN1') 
apple_review = soup.find_all('div',attrs='_3LWZlK')
apple_offs = soup.find_all('div',attrs='_3Ay6Sb')
```

```python
# checking the extracted data :

for name,mrp,dprice,review,offs in zip(apple_name,apple_mrp,apple_discount_price,apple_review,apple_offs):
    print(name.text,mrp.text,dprice.text,review.text,offs.text)
    
```

```python
# Defining headers :

header = ['Brand','Name','MRP','Discount_Price','Reviews','Discount_percentage']
```

```python
# writing all the the extracted data into csv :

with open('Smartphone.csv','w',encoding = 'UTF-8',newline = '') as file:
    writer = csv.writer(file)

# writing header into csv
    writer.writerow(header)
    
# writing apple's data into csv
    for name,mrp,dprice,review,offs in zip(apple_name,apple_mrp,apple_discount_price,apple_review,apple_offs):
        writer.writerow(['Apple',name.text,mrp.text,dprice.text,review.text,offs.text])
# writing samsung's data into csv
    for name,mrp,dprice,review,offs in zip (s_name,s_mrp,s_discount_price,s_review,s_offs):
        writer.writerow(['Samsung',name.text,mrp.text,dprice.text,review.text,offs.text])
# writing oneplus's data into csv
    for name,mrp,dprice,review,offs in zip (op_name,op_mrp,op_discount_price,op_review,op_offs):
        writer.writerow(['OnePluse',name.text,mrp.text,dprice.text,review.text,offs.text])
# writing realme's data into csv
    for name,mrp,dprice,review,offs in zip (rm_name,rm_mrp,rm_discount_price,rm_review,rm_offs):
        writer.writerow(['Realme',name.text,mrp.text,dprice.text,review.text,offs.text])
# writing oppo's data into csv
    for name,mrp,dprice,review,offs in zip (oppo_name,oppo_mrp,oppo_discount_price,oppo_review,oppo_offs):
        writer.writerow(['Oppo',name.text,mrp.text,dprice.text,review.text,offs.text])
# writing vivo's data into csv
    for name,mrp,dprice,review,offs in zip (v_name,v_mrp,v_discount_price,v_review,v_offs):
        writer.writerow(['Vivo',name.text,mrp.text,dprice.text,review.text,offs.text])
# writing google's data into csv
    for name,mrp,dprice,review,offs in zip (g_name,g_mrp,g_discount_price,g_review,g_offs):
        writer.writerow(['Google',name.text,mrp.text,dprice.text,review.text,offs.text])

file.close()
```
